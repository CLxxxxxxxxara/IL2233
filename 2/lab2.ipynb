# 1 Stationarity of AR models

## 1.1

```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

ar1 = np.array([1, -0.8])
ma = np.array([1])
arma_process1 = sm.tsa.ArmaProcess(ar1, ma)
y1 = arma_process1.generate_sample(1000)
plt.plot(y1)
plt.savefig('figure/1_1_1.png')
```

```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

ar2 = np.array([1, 1.1])
ma = np.array([1])
arma_process2 = sm.tsa.ArmaProcess(ar2, ma)
y2 = arma_process2.generate_sample(1000)
plt.plot(y2)
plt.savefig('figure/1_1_2.png')
```

```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

ar3 = np.array([1, -1, 0.5])
ma = np.array([1])
arma_process3 = sm.tsa.ArmaProcess(ar3, ma)
y3 = arma_process3.generate_sample(1000)
plt.plot(y3)
plt.savefig('figure/1_1_3.png')
```

```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

ar4 = np.array([1, -1, -0.5])
ma = np.array([1])
arma_process4 = sm.tsa.ArmaProcess(ar4, ma)
y4 = arma_process4.generate_sample(1000)
plt.plot(y4)
plt.savefig('figure/1_1_4.png')
```

## 1.2 

Time series 1 & 3 are stationary, while 2 & 4 are not

## 1.3

For p = 1: −1 < ϕ1 < 1 => the time series is stationary. So 1 is stationary and 2 is not.
For p = 2: −1 < ϕ2 < 1; ϕ2 + ϕ1 < 1; ϕ2 − ϕ1 < 1 => the time series is stationary. So 3 is stationary and 4 is not.

## 1.4

```python
print(arma_process1.isstationary)
print(arma_process2.isstationary)
print(arma_process3.isstationary)
print(arma_process4.isstationary)
```

```python
from statsmodels.tsa.stattools import adfuller
result_1 = adfuller(y1, autolag='AIC')
print('ADF Statistic: %f' % result_1[0])
print('p-value: %f' % result_1[1])
print('Critical Values:')
for key, value in result_1[4].items():
	print('\t%s: %.3f' % (key, value))
```

```python
result_3 = adfuller(y3, autolag='AIC')
print('ADF Statistic: %f' % result_3[0])
print('p-value: %f' % result_3[1])
print('Critical Values:')
for key, value in result_3[4].items():
	print('\t%s: %.3f' % (key, value))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
def hi_den_box(timeseries,num):
    # Histogram
    plt.hist(timeseries)
    plt.savefig(f'figure/histogram_{num}.png')
    plt.show()

    # Density plot
    sns.kdeplot(timeseries)
    plt.savefig(f'figure/density_{num}.png')
    plt.show()

    # Box plot
    plt.boxplot(timeseries)
    plt.savefig(f'figure/box_{num}.png')
    plt.show()
```

## 2.3

```python
import pandas as pd
def lags(timeseries,num):
    plt.figure()
    series = pd.Series(timeseries)
    pd.plotting.lag_plot(series, lag=1)
    plt.savefig(f'figure/lagone_{num}.png')
    plt.show()
    plt.figure()
    pd.plotting.lag_plot(series, lag=2)
    plt.savefig(f'figure/lagtwo_{num}.png')
    plt.show()
```

```python
lags(y2_1,2_1)
```

```python
lags(y2_2,2_2)
```

```python
lags(y2_3,2_3)
```

```python
lags(y2_4,2_4)
```

# 3 Invertibility, ACF, PACF of MA models

## 3.1

```python
ar = np.array([1])
ma3_1 = np.array([1, -2])
ma3_2 = np.array([1, -0.5])
ma3_3 = np.array([1, -0.8, 0.64])
ma3_4 = np.array([1, -1.25, 1.5625])
arma_process3_1 = sm.tsa.ArmaProcess(ar, ma3_1)
arma_process3_2 = sm.tsa.ArmaProcess(ar, ma3_2)
arma_process3_3 = sm.tsa.ArmaProcess(ar, ma3_3)
arma_process3_4 = sm.tsa.ArmaProcess(ar, ma3_4)
y3_1 = arma_process3_1.generate_sample(1000)
y3_2 = arma_process3_2.generate_sample(1000)
y3_3 = arma_process3_3.generate_sample(1000)
y3_4 = arma_process3_4.generate_sample(1000)
plt.plot(y3_1)
plt.savefig('figure/3_1_1.png')
plt.show()
plt.plot(y3_2)
plt.savefig('figure/3_1_2.png')
plt.show()
plt.plot(y3_3)
plt.savefig('figure/3_1_3.png')
plt.show()
plt.plot(y3_4)
plt.savefig('figure/3_1_4.png')
plt.show()
```

## 3.2

It is not sufficient to determine the invertibility of the series just by its visual inspection of the line plot.

## 3.3

For q = 1, when −1 < θ1 < 1, MA models are invertible. For q = 2, when −1 < θ2 < 1, θ2 + θ1 > −1, θ1 − θ2 < 1, MA models are invertible. Thus, time series 1 and 4 are not invertible, while time series 2 and 3 are invertible.

## 3.4

```python
print(arma_process3_1.isinvertible)
print(arma_process3_2.isinvertible)
print(arma_process3_3.isinvertible)
print(arma_process3_4.isinvertible)
```
